```python
def load_data(url):
    """Loads the dataset from the specified URL."""
    print(f"Loading data from: {url}")
    try:
        df = pd.read_csv(url)
        print("Data loaded successfully.")
        return df
    except Exception as e:
        print(f"Error loading data: {e}")
        return None
```
